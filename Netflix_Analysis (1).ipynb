## **4. Data Cleaning**

```python
# Check for missing values

nf.isnull().sum()
```

```python
nf[nf['title'].isnull()]
```

### Point
- Rows where (title) is NULL and all other Columns are also NULL.
- So i will Drop these rows from data set.

```python
nf.dropna(subset=['title'], inplace=True)
```

```python
nf.isnull().sum()
```

```python
(nf[['cast','country']].isnull().mean() * 100)
```

```python
nf['cast'] = nf['cast'].fillna('Not Available')
nf['country'] = nf['country'].fillna('Not Available')
```

### Handle missing values with 'Cast' & 'Country'

- Identified approximately 9.5% missing values in descriptive columns such as cast and country. These were handled using placeholder imputation to maintain dataset completeness and ensure consistent analysis.

```python
nf.isnull().sum()
```

```python
nf[nf['duration'].isnull()]
```

- duration column has NULL

- but when you look at those rows, the rating column contains values like
"84 min"

```python
mask = nf['duration'].isnull() & nf['rating'].str.contains('min|Season', na=False)

nf.loc[mask, 'duration'] = nf.loc[mask, 'rating']
nf.loc[mask, 'rating'] = 'Not Available'
```

```python
nf.loc[mask, ['rating', 'duration']].head()
```

```python
nf.isnull().sum()
```

```python
# date column is not in Perfect data-type its fix

nf['date_added'] = pd.to_datetime(nf['date_added'], errors='coerce')
```

```python
nf['date_added'].head()
```

```python
nf['release_year'] = nf['release_year'].astype('Int32')
```

```python
nf['release_year'].head()
```

```python
nf['date_added'] = nf['date_added'].fillna('2008')
```

```python
nf['date_added'].isnull().sum()
```

- Missing values in the date_added column were imputed using the earliest available date to ensure temporal consistency where release year information was unreliable.

```python
# Rating columns has 4 null values

most_common_rating = nf['rating'].mode()[0]
nf['rating'] = nf['rating'].fillna(most_common_rating)
```

```python
nf.isnull().sum()
```

# **6. Exploratory Data Analysis (EDA)**

### **6.1 Content Added over Year**

```python
nf['date_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Added Over Years")
plt.show()
```

### - Observation

The number of titles added to Netflix remained minimal until 2015, followed by a sharp increase from 2016 onward. The platform experienced peak content additions between 2018 and 2020, with noticeable fluctuations indicating batch-based releases.

### - Insight

The trend suggests that Netflix underwent rapid expansion after 2016, focusing on scaling its content library and strengthening its competitive position. Post-2020, content additions appear more stable, reflecting a mature and optimized content strategy.

### **6.2 Top Releasing Year**

```python
month_order = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December']
```

```python
nf['month_added'].value_counts().reindex(month_order).plot(kind='bar')
plt.title("Top Month By Added on Netflix")
plt.xlabel('Year')
plt.ylabel('Count')
plt.show()
```

- Observations

January and July have the highest number of titles added on Netflix, while February has the lowest.

Content additions remain fairly consistent from March to October, with a slight drop in November.

December shows an increase, likely due to year-end releases.

- Insights

Netflix strategically releases more content during high-engagement periods such as the start, mid-year, and end of the year.

Lower additions in February and November indicate potential opportunities to improve content distribution during these months.

### **6.3 Movies vs TV Shows**

```python
nf['type'].value_counts()
```

```python
nf['type'].value_counts().plot(kind='pie', autopct='%1.1f%%', figsize=(4,4))
plt.title("Movies vs TV Shows")
plt.ylabel('')
plt.show()
```

### Observation

- Movies constitute approximately 70% of the total content on Netflix.

- TV Shows account for around 30% of the platforms content library.

- The content distribution is clearly skewed towards movies.

### **6.4 Country wise Analysis**

```python
sns.barplot(nf['country'].value_counts().head(10))
plt.title("Top Country with most Content")
plt.ylabel('Content')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

## Observations

- The United States has the highest number of titles on Netflix, followed by India.

- A significant amount of content is marked as Not Available, indicating missing country information.

- Other countries like the UK, Japan, and South Korea contribute comparatively smaller volumes.

## Insights

- Netflixs content library is heavily dominated by the US and Indian markets, reflecting strong production and audience demand.

- Missing country data suggests a need for better data completeness to improve regional content analysis.

### **6.5 Rating Distribution**

```python
rating_dis = nf['rating'].value_counts()
```

```python
sns.barplot(x=rating_dis.index, y=rating_dis.values , color='red')
plt.title("Rating Distribution")
plt.xlabel('Ratings Type')
plt.ylabel('Count of Rating')
plt.xticks(rotation=60)
plt.tight_layout()
plt.show()
```

### Observation

The rating distribution shows that TV-MA and TV-14 dominate Netflixs content library, while family-friendly ratings form a smaller proportion.

### Insight

This indicates Netflixs strategic focus on mature and teen audiences, aligning content offerings with its primary subscriber demographics.

### **6.6 Director's wise Movies/TV-shows**

```python
nf['director'].value_counts().sort_values(ascending=False).head(10)
```

### Observation

A significant portion of Netflix titles lack director information, while the remaining catalog shows a wide distribution of directors with relatively small title counts.
